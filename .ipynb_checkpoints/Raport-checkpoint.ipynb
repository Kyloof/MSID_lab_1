## 2. Extracting basic statistics from the file

```python
import pandas as pd
from load_data import load_data

def generate_statistics_from_file(file_path, output_path):
    df = load_data(file_path)

    # Define categorical columns explicitly (including Year)
    categorical_columns = ["Year"] + list(df.select_dtypes(exclude=['int64', 'float64']).columns)
    numeric_columns = [col for col in df.select_dtypes(include=['int64', 'float64']).columns if col not in categorical_columns]

    # Dictionary to store numeric statistics
    numeric_stats = {}
    for attr in numeric_columns:
        missing_values = df[attr].isnull().sum()
        stats = df[attr].describe(percentiles=[0.05, 0.95]).to_dict()
        stats["missing_values"] = missing_values
        numeric_stats[attr] = stats
    
    # Convert numeric statistics dictionary to DataFrame
    numeric_df = pd.DataFrame(numeric_stats).transpose()
    numeric_df.insert(0, 'attribute', numeric_df.index)
    
    # List to store categorical statistics
    categorical_stats = []
    for attr in categorical_columns:
        missing_values = df[attr].isnull().sum()
        unique_classes = df[attr].nunique()
        class_proportions = df[attr].value_counts(normalize=True).to_dict()
        
        categorical_stats.append({
            'attribute': attr,
            'unique_classes_amount': unique_classes,
            'missing_values': missing_values,
            'class_proportions': class_proportions
        })
    
    # Convert categorical statistics list to DataFrame
    categorical_df = pd.DataFrame(categorical_stats)
    
    # Save outputs to CSV
    numeric_df.to_csv(output_path.replace('.csv', '_numeric.csv'), index=False)
    categorical_df.to_csv(output_path.replace('.csv', '_categorical.csv'), index=False)

# Example usage
generate_statistics_from_file("UltimateClassicRock.csv", "UltimateClassicRockOutput.csv")

    
```

## 3. BoxPlot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.catplot(data=df, x="Year", y="Tempo", kind="box", height=9, aspect=2)
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'  
)
plt.title("Distribution of Tempo Across Different Years")

plt.show()
```

We can observe that since the 60's the average tempo of the song stays nearly the same, however with the evolution of rock music the standard deviation has gotten larger in the 70's, 80's

```python


def plot_tempo_errorbars(error_type='sd', **kws):
    tempo = df['Tempo']  
    plt.figure(figsize=(8, 6))
    sns.pointplot(x=[0], y=tempo, errorbar=error_type, **kws, capsize=0.2)
    
    plt.title(f'Tempo: Error Bars ({error_type})')
    plt.ylabel('Tempo')
    plt.xlabel('Tempo (Index)')
    
    # Display the plot
    plt.show()

# Example usage: plot Tempo with standard deviation (SD) error bars
plot_tempo_errorbars(error_type='sd')
```
